Hoeveel strings zijn 'nice'? Een string is nice als:
It contains at least three vowels (aeiou only), like aei, xazegov, or aeiouaeiouaeiou.
It contains at least one letter that appears twice in a row, like xx, abcdde (dd), or aabbccdd (aa, bb, cc, or dd).
It does not contain the strings ab, cd, pq, or xy, even if they are part of one of the other requirements.

```python
import re

with open('input.txt', 'r') as file:
    data = file.readlines()
```

```python
#deze functie checkt of een text 3x een vowel bevat
#returned de hele string als waar, anders returned het een lege array
def multi_vowel_words(text):
   pattern = r"\b(?:\w*[aeiouAEIOU]){3}\w*"
   result = re.findall(pattern, text)
   return result

#deze functie checkt of een text tenminste 1x een letter heeft dat twee keer achter elkaar komt
def twice_in_a_row(text):
   current = "loremipsum"
   result = False
   for c in text:
      if c == current:
         result = True
      else:
         current = c
   return result

#deze functie checkt of de string niet ab, cd, pq of xy bevat
def contains_substring(text):
   result = False
   if "ab" in text:
      result = True
   elif "cd" in text:
      result = True
   elif "pq" in text:
      result = True
   elif "xy" in text:
      result = True
   return result
```

```python
mylist = []
mylist_twice = []
mylist_three = []

#Voeg alle woorden die minimaal drie vowels hebben toe aan mylist
for string in data:
  test = multi_vowel_words(string)
  if test != []:
    mylist += test

#voeg alle woorden waarvan
for string in mylist:
  test = twice_in_a_row(string)
  if test == True:
    mylist_twice.append(string)

for string in mylist_twice:
  test = contains_substring(string)
  if test == False:
    mylist_three.append(string)

print(len(mylist_three))
```
